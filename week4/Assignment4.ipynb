```python

import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("C:/Users/sudob/OneDrive/Desktop/week4/tigre.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image)
plt.title("Original Image")
plt.axis("off")
plt.show()
```

```python

gamma = 2.0

normalized = image / 255.0
gamma_corrected = np.power(normalized, gamma)
gamma_corrected = np.uint8(gamma_corrected * 255)

plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.imshow(image)
plt.title("Original")
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(gamma_corrected)
plt.title("Gamma Corrected")
plt.axis('off')

plt.show()
```

```python

r = np.linspace(0,1,256)

plt.figure(figsize=(8,6))

for gamma in [0.2, 0.5, 1.0, 2.0, 5.0]:
    s = r ** gamma
    plt.plot(r,s,label="gamma="+str(gamma))

plt.title("Gamma Curves")
plt.xlabel("Input Intensity")
plt.ylabel("Output Intensity")
plt.legend()
plt.grid()

plt.show()
```

```python

plt.figure(figsize=(15,5))

gamma_list = [0.5, 1.0, 2.0]

for i in range(len(gamma_list)):

    gamma = gamma_list[i]

    normalized = image / 255.0
    corrected = np.power(normalized, gamma)
    corrected = np.uint8(corrected * 255)

    plt.subplot(1,3,i+1)
    plt.imshow(corrected)
    plt.title("Gamma="+str(gamma))
    plt.axis('off')

plt.show()
```

```python

negative = 255 - image

plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.imshow(image)
plt.title("Original")
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(negative)
plt.title("Negative")
plt.axis('off')

plt.show()
```

```python

A = 100
B = 180

mapped = image.copy()

mask = (mapped >= A) & (mapped <= B)
mapped[mask] = 255

plt.imshow(mapped)
plt.title("Intensity Mapping")
plt.axis('off')
plt.show()
```

```python

gamma = 1.5

normalized = image / 255.0
gamma_img = np.power(normalized, gamma)
gamma_img = np.uint8(gamma_img * 255)

A = 100
B = 200

enhanced = gamma_img.copy()
mask = (enhanced >= A) & (enhanced <= B)
enhanced[mask] = 255

plt.imshow(enhanced)
plt.title("Enhanced Image")
plt.axis('off')
plt.show()
```

```python
gamma1 = 0.5
gamma2 = 2.0
gamma3 = 1.0
norm = image / 255.0

gamma_low = np.power(norm, gamma1)
gamma_low = np.uint8(gamma_low * 255)

gamma_high = np.power(norm, gamma2)
gamma_high = np.uint8(gamma_high * 255)

negative_final = 255 - image

A = 100
B = 180

mapped_final = image.copy()
mask = (mapped_final >= A) & (mapped_final <= B)
mapped_final[mask] = 255

plt.figure(figsize=(15,10))

plt.subplot(2,3,1)
plt.imshow(image)
plt.title("Original Image")
plt.axis('off')

plt.subplot(2,3,2)
plt.imshow(negative_final)
plt.title("Negative Image")
plt.axis('off')

plt.subplot(2,3,3)
plt.imshow(mapped_final)
plt.title("Intensity Mapping")
plt.axis('off')

plt.subplot(2,3,4)
plt.imshow(gamma_low)
plt.title("Gamma = 0.5")
plt.axis('off')

plt.subplot(2,3,5)
plt.imshow(gamma_high)
plt.title("Gamma = 2.0")
plt.axis('off')

plt.subplot(2,3,6)
plt.imshow(gamma_high)
plt.title("Gamma = 1.0")
plt.axis('off')
plt.tight_layout()
plt.show()
```
