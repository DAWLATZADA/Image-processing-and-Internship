```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("C:/Users/sudob/OneDrive/Desktop/week4/image.jpg", cv2.IMREAD_GRAYSCALE)

plt.imshow(image, cmap='gray')
plt.title("Original Image")
plt.axis("off")
```

```python
def gamma_correction(I_in, c, gamma):

    # Normalize image
    I_norm = I_in.astype('float32') / 255

    # Apply power law
    I_gamma = c * (I_norm ** gamma)

    return I_gamma
```

```python
gamma_img = gamma_correction(image, 1, 2)

plt.imshow(gamma_img, cmap='gray')
plt.title("Gamma Corrected Image")
plt.axis("off")
```

```python
save_img = (gamma_img * 255).astype(np.uint8)
cv2.imwrite("gamma_output.jpg", save_img)
```

```python
def contrast_stretching(I_in, smin, smax):

    rmin = np.min(I_in)
    rmax = np.max(I_in)

    output = ((smax - smin) / (rmax - rmin)) * (I_in - rmin) + smin

    return output
```

```python
contrast_img = contrast_stretching(image, 0, 255)

plt.imshow(contrast_img, cmap='gray')
plt.title("Contrast Stretched Image")
plt.axis("off")
```

```python
image_norm = image.astype('float32') / 255

contrast_norm = contrast_stretching(image_norm, 0, 1)

plt.imshow(contrast_norm, cmap='gray')
plt.title("Normalized Contrast Stretching")
plt.axis("off")
```
